# Javascript in Databricks Notebooks

This notebook illustrates the various methods using which you can use HTML and Javascript inside Databricks notebooks

<strong>
You can use HTML tags inside markdown cells. This example uses the &ltstrong&gt tag
</strong>

```
%%html
<strong>
You can use the %%html cell magic
</strong>
```

```
%python
displayHTML("""
<strong>
You can use the displayHTML function in a python cell
</strong>
""")
```

```
%%html
<div id="output"></div>
<script>
document.querySelector('#output').innerHTML = 'You can put Javascript inside %%html';
</script>
```

```
%python
script = """<script>
document.querySelector('#output').innerHTML = 'Javascript passed into displayHTML';
</script>"""

displayHTML(f"""
<div id="output"></div>
{script}
""")
```
